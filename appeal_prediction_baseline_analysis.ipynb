```python
# ==== 0) Imports ====
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, accuracy_score, precision_score, recall_score, roc_curve
import matplotlib.pyplot as plt

# ==== 1) Get your DataFrame ====
# If you already have df from your upload step, keep it.
# Otherwise, uncomment one of these:

# from google.colab import files
# uploaded = files.upload()  # pick appeal_data.xlsx or csv
# df = pd.read_excel('appeal_data.xlsx')  # or pd.read_csv('appeal_data.csv')

# Quick sanity check
print("Columns:", list(df.columns))
display(df.head(10))

# ==== 2) Target setup ====
# Well predict whether an account appeals (1) vs not (0).
# If APPEALNO is numeric: treat >0 as appealed. If it's missing, fill 0.
df = df.copy()
if "APPEALNO" not in df.columns:
    raise ValueError("Expected a column named 'APPEALNO' for appeal indicator.")

df["APPEALNO"] = pd.to_numeric(df["APPEALNO"], errors="coerce").fillna(0)
df["appealed"] = (df["APPEALNO"] > 0).astype(int)

# ==== 3) Basic cleaning for typical assessor exports ====
# Treat 0/9999 years as missing
for col in ["AYB", "EYB"]:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")
        df[col] = df[col].replace({0: np.nan, 9999: np.nan})

# Coerce known numeric columns (wont break if missing)
for col in ["ECONOMICAREACODE", "BLTASCODE", "AYB", "EYB", "PRIOR", "CURRENT"]:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

# ==== 4) Feature selection ====
# Exclude identifiers and the original APPEALNO/label
drop_cols = {"ACCOUNTNO", "APPEALNO", "appealed"}  # add any other pure IDs here
feature_cols = [c for c in df.columns if c not in drop_cols]

# Separate numeric vs categorical
numeric_cols = [c for c in feature_cols if pd.api.types.is_numeric_dtype(df[c])]
categorical_cols = [c for c in feature_cols if not pd.api.types.is_numeric_dtype(df[c])]

print("Numeric features:", numeric_cols)
print("Categorical features:", categorical_cols)

X = df[feature_cols]
y = df["appealed"]

# ==== 5) Train/test split ====
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

# ==== 6) Preprocess + model ====
numeric_pipe = Pipeline(steps=[
    ("impute", SimpleImputer(strategy="median")),
    ("scale", StandardScaler())
])

categorical_pipe = Pipeline(steps=[
    ("impute", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])

preprocess = ColumnTransformer(
    transformers=[
        ("num", numeric_pipe, numeric_cols),
        ("cat", categorical_pipe, categorical_cols)
    ],
    remainder="drop"
)

model = LogisticRegression(max_iter=2000)

clf = Pipeline(steps=[
    ("prep", preprocess),
    ("model", model)
])

# ==== 7) Train ====
clf.fit(X_train, y_train)

# ==== 8) Evaluate ====
probs = clf.predict_proba(X_test)[:, 1]
preds = (probs >= 0.5).astype(int)

roc_auc = roc_auc_score(y_test, probs)
acc = accuracy_score(y_test, preds)
prec = precision_score(y_test, preds, zero_division=0)
rec = recall_score(y_test, preds, zero_division=0)

print("\n=== Test Metrics ===")
print(f"ROC AUC : {roc_auc:.3f}")
print(f"Accuracy: {acc:.3f}")
print(f"Precision: {prec:.3f}")
print(f"Recall   : {rec:.3f}")

# ROC curve
fpr, tpr, _ = roc_curve(y_test, probs)
plt.figure()
plt.plot(fpr, tpr, label=f"LogReg (AUC={roc_auc:.3f})")
plt.plot([0,1],[0,1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()

# ==== 9) Score ALL accounts and export top prospects ====
all_probs = clf.predict_proba(X)[:, 1]
out = df[["ACCOUNTNO"]].copy() if "ACCOUNTNO" in df.columns else pd.DataFrame(index=df.index)
out["appeal_probability"] = all_probs
out["appeal_prediction"] = (all_probs >= 0.5).astype(int)

# Top 50 most likely to appeal
top_n = out.sort_values("appeal_probability", ascending=False).head(50)
display(top_n.head(10))

# Save files
out_path_all = "appeal_predictions_all.csv"
out_path_top = "appeal_predictions_top50.csv"
out.to_csv(out_path_all, index=False)
top_n.to_csv(out_path_top, index=False)
print(f"\nSaved: {out_path_all} and {out_path_top}")
```

```python
# ===========================
# Try to match VS Code setup
# ===========================
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    roc_auc_score, accuracy_score, precision_score, recall_score,
    roc_curve, f1_score, confusion_matrix
)
import matplotlib.pyplot as plt

# --- 0) Config ---
RANDOM_STATE = 42     # <- set this to the same seed you used in VS Code
TEST_SIZE = 0.25

# If df isn't defined (e.g., after a reset), try to load from uploaded file
if 'df' not in globals():
    try:
        df = pd.read_excel('appeal_data.xlsx')
    except Exception:
        df = pd.read_csv('appeal_data.csv')

df = df.copy()

# --- 1) Target variable: match VS Code logic (appealed if APPEALNO > 0) ---
if "APPEALNO" not in df.columns:
    raise ValueError("Column 'APPEALNO' not found. Make sure your dataset has it.")
df["APPEALNO"] = pd.to_numeric(df["APPEALNO"], errors="coerce").fillna(0)
df["appealed"] = (df["APPEALNO"] > 0).astype(int)

# Inspect imbalance
pos_rate = df["appealed"].mean()
print(f"Appeal positive rate: {pos_rate:.4%} (class imbalance check)")

# --- 2) Light, transparent feature engineering (safe defaults) ---
# Convert known numeric-like columns
for col in ["ECONOMICAREACODE", "BLTASCODE", "AYB", "EYB", "PRIOR", "CURRENT"]:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

# Treat obvious placeholders
for col in ["AYB", "EYB"]:
    if col in df.columns:
        df[col] = df[col].replace({0: np.nan, 9999: np.nan})

# Add valuation deltas if PRIOR/CURRENT exist
if {"PRIOR", "CURRENT"}.issubset(df.columns):
    df["valuation_change"] = df["CURRENT"] - df["PRIOR"]
    # Avoid div-by-zero; use safe percent change
    denom = df["PRIOR"].replace({0: np.nan})
    df["valuation_pct_change"] = (df["CURRENT"] - df["PRIOR"]) / denom
    df["valuation_pct_change"] = df["valuation_pct_change"].replace([np.inf, -np.inf], np.nan)

# Optional: simple age feature if AYB available
if "AYB" in df.columns:
    df["building_age"] = 2025 - df["AYB"]  # adjust year if your VS Code used a different reference

# --- 3) Feature set ---
drop_cols = {"ACCOUNTNO", "APPEALNO", "appealed"}  # add other pure IDs if present
feature_cols = [c for c in df.columns if c not in drop_cols]

# Split numeric/categorical by dtype
numeric_cols = [c for c in feature_cols if pd.api.types.is_numeric_dtype(df[c])]
categorical_cols = [c for c in feature_cols if not pd.api.types.is_numeric_dtype(df[c])]

print("\nNumeric features:", numeric_cols)
print("Categorical features:", categorical_cols)

X = df[feature_cols]
y = df["appealed"]

# --- 4) Train/test split (match seed + stratify) ---
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE, stratify=y
)

# --- 5) Preprocess + model (balanced class weights) ---
num_pipe = Pipeline([
    ("impute", SimpleImputer(strategy="median")),
    ("scale", StandardScaler())
])

cat_pipe = Pipeline([
    ("impute", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])

preprocess = ColumnTransformer(
    transformers=[
        ("num", num_pipe, numeric_cols),
        ("cat", cat_pipe, categorical_cols)
    ],
    remainder="drop"
)

logreg = LogisticRegression(
    max_iter=5000,
    class_weight="balanced",   # <- helps with rare appeals
    solver="liblinear",        # stable with class_weight; you can use 'lbfgs' too
    random_state=RANDOM_STATE
)

pipe = Pipeline([
    ("prep", preprocess),
    ("clf", logreg)
])

# --- 6) Train ---
pipe.fit(X_train, y_train)

# --- 7) Evaluate at default 0.5 and also at an optimized threshold ---
probs = pipe.predict_proba(X_test)[:, 1]

def report_at_threshold(th, name):
    preds = (probs >= th).astype(int)
    roc_auc = roc_auc_score(y_test, probs)
    acc = accuracy_score(y_test, preds)
    prec = precision_score(y_test, preds, zero_division=0)
    rec = recall_score(y_test, preds, zero_division=0)
    f1 = f1_score(y_test, preds, zero_division=0)
    cm = confusion_matrix(y_test, preds)
    print(f"\n=== {name} ===")
    print(f"Threshold : {th:.3f}")
    print(f"ROC AUC   : {roc_auc:.3f}")
    print(f"Accuracy  : {acc:.3f}")
    print(f"Precision : {prec:.3f}")
    print(f"Recall    : {rec:.3f}")
    print(f"F1        : {f1:.3f}")
    print("Confusion matrix [TN FP; FN TP]:")
    print(cm)

# Default 0.5
report_at_threshold(0.5, "Metrics @ 0.5 cutoff")

# Choose an operating point: maximize Youden's J on ROC (TPR - FPR)
fpr, tpr, thresholds = roc_curve(y_test, probs)
youden_j = tpr - fpr
best_idx = np.argmax(youden_j)
best_th = thresholds[best_idx]
report_at_threshold(best_th, "Metrics @ Youden-J optimal cutoff")

# Plot ROC
plt.figure()
plt.plot(fpr, tpr, label=f"LogReg (AUC={roc_auc_score(y_test, probs):.3f})")
plt.plot([0,1],[0,1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()

# --- 8) Score everyone and produce a unique-accounts Top 50 (if ACCOUNTNO exists) ---
all_probs = pipe.predict_proba(X)[:, 1]
scored = pd.DataFrame(index=df.index)
if "ACCOUNTNO" in df.columns:
    scored["ACCOUNTNO"] = df["ACCOUNTNO"]
scored["appeal_probability"] = all_probs
scored["appeal_prediction_0p5"] = (all_probs >= 0.5).astype(int)
scored["appeal_prediction_best"] = (all_probs >= best_th).astype(int)

# Collapse to unique accounts by max probability
if "ACCOUNTNO" in scored.columns:
    top50_unique = (
        scored.groupby("ACCOUNTNO", as_index=False)
              .appeal_probability.max()
              .sort_values("appeal_probability", ascending=False)
              .head(50)
    )
    print("\nTop 10 (unique accounts) by appeal probability:")
    display(top50_unique.head(10))
else:
    print("\nACCOUNTNO not found; showing top 10 rows instead:")
    display(scored.sort_values("appeal_probability", ascending=False).head(10))

# Save (optional)
scored.to_csv("appeal_predictions_all_balanced.csv", index=False)
if "ACCOUNTNO" in scored.columns:
    top50_unique.to_csv("appeal_predictions_top50_unique_balanced.csv", index=False)
print("\nSaved: appeal_predictions_all_balanced.csv and (if applicable) appeal_predictions_top50_unique_balanced.csv")
```

# Appeal Prediction Model  Summary & Recommendations

## Key Findings
- **Appeals are rare**: Only **2.57%** of properties in the dataset have an appeal.
- **Performance**:
  - **ROC AUC**: 0.63  There is some predictive signal, but its weak  appeals and non-appeals look similar to the model.
  - **Youden-J optimal threshold (0.510)**:
    - Accuracy: 0.641
    - Precision: 0.041 (many false positives)
    - Recall: 0.588 (finds most appeals)
- **High recall, low precision**: The model catches many appeals but also flags a lot of non-appeals.

---

## How to Improve Results

### 1. Add Better Features
- Property change trends (valuation %, comparison to area average).
- Time since last appeal or major value change.
- Property details (type, size, zoning, units).
- Relative comparisons to similar properties nearby.

### 2. Improve Data Quality
- Fix inconsistent category names.
- Replace placeholder values like `0` or `9999` with nulls.
- Fill in missing build/eff. year data where possible.

### 3. Handle Class Imbalance
- Oversample appeals or undersample non-appeals.
- Keep `class_weight="balanced"` in models.

### 4. Try Other Models
- Use Random Forest or Gradient Boosting to capture complex patterns.
- Compare against logistic regression baseline.

### 5. Choose a Smarter Threshold
- Use metrics to pick a cutoff that fits business needs:
  - Lower threshold  more recall (catch more appeals).
  - Higher threshold  more precision (fewer false positives).
